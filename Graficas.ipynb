```
import matplotlib.pyplot as plt

# =========================
# 1. Gráfica de Barras
# =========================
meses = ["Enero", "Febrero", "Marzo", "Abril"]
ventas = [50, 75, 60, 90]

plt.figure(figsize=(6,4))
plt.bar(meses, ventas, color="skyblue")
plt.title("Ventas Mensuales")
plt.xlabel("Meses")
plt.ylabel("Ventas")
plt.show()


# =========================
# 2. Gráfica de Línea
# =========================
dias = ["Lunes", "Martes", "Miércoles", "Jueves", "Viernes"]
temperatura = [20, 22, 19, 24, 23]

plt.figure(figsize=(6,4))
plt.plot(dias, temperatura, marker="o", color="green")
plt.title("Temperatura Semanal")
plt.xlabel("Días")
plt.ylabel("Temperatura (°C)")
plt.grid(True)
plt.show()


# =========================
# 3. Gráfica de Pastel
# =========================
categorias = ["Renta", "Comida", "Transporte", "Entretenimiento", "Ahorro"]
porcentajes = [40, 25, 15, 10, 10]

plt.figure(figsize=(6,6))
plt.pie(porcentajes, labels=categorias, autopct="%1.1f%%", startangle=90)
plt.title("Distribución de Gastos Mensuales")
plt.show()
```
